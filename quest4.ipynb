```python
#Approximating sine function upto 4 decimal places and plotting a graph between Modulus of errors and iterations 

import math
import matplotlib.pyplot  as plt 
def fact_rial(num) :#Running the factorial statement
   if num < 0:
      print("Sorry, factorial does not exist for negative numbers")
   else:
      i = 0
      prod = 1
      j = 1
      while i < num :
         prod = prod * j
         j = j + 1
         i = i + 1
      return prod

def sin( x, n):#defining sine function
    sine = 0
    for i in range(n):
        sign = (-1)**i
        y = x*(math.pi/180)#Converting degrees to radians
        sine += ((y**(2.0*i+1))/fact_rial(2*i+1))*sign
    return sine
 

x=1.9
list1=[]
list2=[]
for i in range(6):
    list1.append(i)
    list2.append(round(abs(math.sin(x)-sin(x,i)),4))#making a list of plotting variables
    
print(list1)
print(list2)

plt.xlabel("Iteration Numbers(n)")
plt.ylabel("Modulus of errors")
plt.plot(list1,list2)


```
